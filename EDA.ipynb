### release between 2010 to 2019

```python
df_filter = df[(df['release_year'] > 2010) & (df['release_year'] < 2019)]
df_filter = df_filter.head(2000)
df_filter
```

```python
plt.figure(figsize = (10,8))

mdata = sns.countplot(data=df_filter, x='rating', hue='type')
plt.title('Content Type available according to ratings')
plt.xlabel('Rating')
mdata.legend(title = 'legend title', loc='center right',bbox_to_anchor=(0.95,0.9))
plt.show()
```
